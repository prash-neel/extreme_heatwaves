```python
#use xarray package to create new netcdfs for pattern correlation
#compare 2020 siberian heatwave to reanalysis data and find pattern corr
#Low res version - reduce resolution of reanalysis to allow like for like comparison to model data patter correlation

##IMPORT PACKAGES
from sewar.full_ref import mse, rmse, psnr, uqi, ssim, ergas, scc, rase, sam, msssim, vifp
import cartopy.feature as cfeature ## for lakes/borders etc
countries_50m = cfeature.NaturalEarthFeature('cultural','admin_0_countries','50m',edgecolor='k',facecolor='none')
import cartopy.crs as ccrs
import numpy as np
from netCDF4 import Dataset  
import xarray as xr
import datetime as dt
import pandas as pd

import matplotlib.pyplot as plt
import math
import matplotlib.ticker as mticker
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import os
import iris
from iris.experimental.equalise_cubes import equalise_attributes
from iris.util import unify_time_units
import iris.coord_categorisation
import iris.plot as iplt
import iris.quickplot as qplt
import iris.analysis.stats
import pandas as pd
import seaborn as sns
import glob
import cf_units as unit

crs = ccrs.PlateCarree()

import warnings
warnings.filterwarnings("ignore")

def sorted(cube, coord):
    coord_to_sort = cube.coord(coord)
    assert coord_to_sort.ndim == 1, 'One dim coords only please.'
    dim, = cube.coord_dims(coord_to_sort)
    index = [slice(None)] * cube.ndim
    index[dim] = np.argsort(coord_to_sort.points)
    return cube[tuple(index)]
```

```python
#LOWRES resample cubes to make smaller and match lat/longs from model data
sample_points = [('longitude', longitude_lowRES.points),
                 ('latitude',  latitude_lowRES.points)]
resampled_cube1 = cube1_sorted.interpolate(sample_points, iris.analysis.Linear())
resampled_cube2 = cube2_sorted.interpolate(sample_points, iris.analysis.Linear())

cosinelat_weights=iris.analysis.cartography.cosine_latitude_weights(resampled_cube1)
```

```python
#calculate correlation

corr = np.zeros(len(dt_time2))
for i in np.arange(len(dt_time2)):
    correlation = iris.analysis.stats.pearsonr(resampled_cube1, resampled_cube2[i], weights=cosinelat_weights)
    corr[i] = correlation.data
    
```

```python
#load datasets to apply weighting for RMSE
unweighted_zHW = resampled_cube1.data

#model
unweighted_zSIB = resampled_cube2.data

zSIB = unweighted_zSIB * cosinelat_weights
zHW = unweighted_zHW * cosinelat_weights

#corr2= np.corrcoef(z[1],z[1])
print('Done!')

print(zHW.shape)
print(zSIB.shape)
print(unweighted_zSIB.shape)
```

```python
#Calculate RMSE and spatial corr

SCC = np.zeros(len(dt_time2))
RMSE = np.zeros(len(dt_time2))
for i in np.arange(len(dt_time2)):
    SCC[i] = scc(zHW,zSIB[i])
    RMSE[i] = rmse(zHW,zSIB[i])
```

```python
corr_series = pd.DataFrame({'dates':dt_time2,'p_corr': corr,'SCC': SCC, 'RMSE': RMSE}, columns=['dates','p_corr','SCC', 'RMSE'])
corr_series
```
